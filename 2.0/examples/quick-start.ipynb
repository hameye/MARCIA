## Import Required Libraries

```python
import marcia as mc
```

## Load Data

```python
data = mc.load('Data/lead_ore','.bmp', True)
```

```python
mc.hist(data, 'Pb')
```

## Load Classification Table

```python
mask = mc.load_mask('Data/Mask.xlsx')
```

```python
mineral_cube = mc.mineralcube_creation(data,mask)
```

## See classification

```python
mc.plot_minerals(mineral_cube)
```

## Check classification accuracy

```python
mc.hist(mineral_cube, 'Galene')
```

## Export mask to image

Masks can be save as raw tiff images.

```python
mc.save(mineral_cube, 'Galene', raw=True)
```
